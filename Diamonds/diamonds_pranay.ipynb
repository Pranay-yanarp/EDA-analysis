# EDA analysis of Diamonds dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
df=sns.load_dataset('diamonds')
```

Variable | Description | Values
--------:|:--------:|:---------
price | price in US dollars | $326-$18,823
carat | weight of the diamond | 0.2-5.01
cut | quality of the cut | Fair, Good, Very Good, Premium, Ideal
color | diamond color | J (worst) to D (best)
clarity | measurement of how clear the diamond is | I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best)
x | length in mm | 0-10.74
y | width in mm | 0-58.9
z | depth in mm | 0-31.8
depth | total depth percentage | 43-79
table | width of top of diamond relative to widest point | 43-95

```python


e=[]
i=[]
j=[]
h=[]
f=[]
g=[]
d=[]

df_mean=pd.DataFrame(columns=['cut','E','I','J','H','F','G','D'])
df_mean['cut']=['Ideal', 'Premium', 'Very Good','Good', 'Fair']
cuts=['Ideal', 'Premium', 'Very Good','Good', 'Fair']


for z in cuts:
    e.append(round(np.mean(df[((df['cut']==z)&(df['color']=='E'))]['price']),3))
    i.append(round(np.mean(df[((df['cut']==z)&(df['color']=='I'))]['price']),3))
    j.append(round(np.mean(df[((df['cut']==z)&(df['color']=='J'))]['price']),3))
    h.append(round(np.mean(df[((df['cut']==z)&(df['color']=='H'))]['price']),3))
    f.append(round(np.mean(df[((df['cut']==z)&(df['color']=='F'))]['price']),3))
    g.append(round(np.mean(df[((df['cut']==z)&(df['color']=='G'))]['price']),3))
    d.append(round(np.mean(df[((df['cut']==z)&(df['color']=='D'))]['price']),3))

df_mean['E']=e
df_mean['I']=i
df_mean['J']=j
df_mean['H']=h
df_mean['F']=f
df_mean['G']=g
df_mean['D']=d

df_mean_1=pd.DataFrame(columns=['cut','E','I','J','H','F','G','D'])

df_mean_1['cut']=['Maximum','Minimum']

e1=[]
i1=[]
j1=[]
h1=[]
f1=[]
g1=[]
d1=[]


e1.append(df_mean.loc[df_mean['E'].argmax(),'cut'])
i1.append(df_mean.loc[df_mean['I'].argmax(),'cut'])
j1.append(df_mean.loc[df_mean['J'].argmax(),'cut'])
h1.append(df_mean.loc[df_mean['H'].argmax(),'cut'])
f1.append(df_mean.loc[df_mean['F'].argmax(),'cut'])
g1.append(df_mean.loc[df_mean['G'].argmax(),'cut'])
d1.append(df_mean.loc[df_mean['D'].argmax(),'cut'])


e1.append(df_mean.loc[df_mean['E'].argmin(),'cut'])
i1.append(df_mean.loc[df_mean['I'].argmin(),'cut'])
j1.append(df_mean.loc[df_mean['J'].argmin(),'cut'])
h1.append(df_mean.loc[df_mean['H'].argmin(),'cut'])
f1.append(df_mean.loc[df_mean['F'].argmin(),'cut'])
g1.append(df_mean.loc[df_mean['G'].argmin(),'cut'])
d1.append(df_mean.loc[df_mean['D'].argmin(),'cut'])


df_mean_1['E']=e1
df_mean_1['I']=i1
df_mean_1['J']=j1
df_mean_1['H']=h1
df_mean_1['F']=f1
df_mean_1['G']=g1
df_mean_1['D']=d1

df_new=df_mean.append(df_mean_1,ignore_index=True)

max12=[]
min12=[]

# df_mean.index=df_mean['cut']
for l in range(len(df_mean['cut'])):
    max1=np.max(df_mean.iloc[l,1:])
    for l1 in df_mean.columns[1:]:
        if df_mean.loc[l,l1]==max1:
            max12.append(l1)
    min1=np.min(df_mean.iloc[l,1:])
    for l1 in df_mean.columns[1:]:
        if df_mean.loc[l,l1]==min1:
            min12.append(l1)

max12.append(np.nan)
max12.append(np.nan)
min12.append(np.nan)
min12.append(np.nan)
df_new['Maximum']=max12
df_new['Minimum']=min12

print('Mean price value of diamonds of various cut and color')
print(tabulate(df_new, headers='keys', tablefmt='fancy_grid'))

```
